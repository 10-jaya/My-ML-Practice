```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
data=pd.read_csv("Assignment4_SalaryData.csv")
data
```

```python
style.use("fivethirtyeight")
plt.plot(data.YearsExperience,data.Salary)
plt.title("Salary VS Experience", fontname="Arial", fontsize=24)
plt.xlabel("Years Experience", fontname="Arial", fontsize=18)
plt.ylabel("Salary", fontname="Arial", fontsize=18)
plt.show()
```
